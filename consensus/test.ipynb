```python
import methods
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
def plot_results(f_err, cons_err, dist, method=None, graph_model=None):

    fig, ax = plt.subplots(1, 3, figsize=(15, 4))

    ax[0].plot(f_err, color='red', label=method)
    ax[0].legend()
    #ax[0].yscale('log')
    ax[0].set_title("Function")
    ax[0].set_xlabel("Iteration number")
    ax[0].set_ylabel("Error")
    ax[0].grid(alpha=0.1)

    ax[1].plot(cons_err, color='green', label=method)
    ax[1].legend()
    ax[1].set_yscale('log')
    ax[1].set_title("Consensus")
    ax[1].set_xlabel("Iteration number")
    ax[1].set_ylabel("Squared rror")
    ax[1].grid(alpha=0.1)

    ax[2].plot(dist, color='blue', label=method)
    ax[2].legend()
    ax[2].set_yscale('log')
    ax[2].set_title("Primal variable")
    ax[2].set_xlabel("Iteration number")
    ax[2].set_ylabel("Squared error")
    ax[2].grid(alpha=0.1)

    plt.suptitle(graph_model, fontsize=14)
```

```python
def plot_results_comparison(f_err1, f_err2, cons_err1, cons_err2, dist1, dist2, method1=None, method2=None, graph_model=None):

    fig, ax = plt.subplots(1, 3, figsize=(15, 4))

    ax[0].plot(f_err1, color='red', label=method1)
    ax[0].plot(f_err2, color='red', linestyle='dashed', label=method2)
    ax[0].legend()
    #ax[0].yscale('log')
    ax[0].set_title("Function")
    ax[0].set_xlabel("Iteration number")
    ax[0].set_ylabel("Error")
    ax[0].grid(alpha=0.1)

    ax[1].plot(cons_err1, color='green', label=method1)
    ax[1].plot(cons_err1, color='green', linestyle='dashed', label=method2)
    ax[1].legend()
    ax[1].set_yscale('log')
    ax[1].set_title("Consensus")
    ax[1].set_xlabel("Iteration number")
    ax[1].set_ylabel("Squared rror")
    ax[1].grid(alpha=0.1)

    ax[2].plot(dist1, color='blue', label=method1)
    ax[2].plot(dist2, color='blue', linestyle='dashed', label=method2)
    ax[2].legend()
    ax[2].set_yscale('log')
    ax[2].set_title("Primal variable")
    ax[2].set_xlabel("Iteration number")
    ax[2].set_ylabel("Squared error")
    ax[2].grid(alpha=0.1)

    plt.suptitle(graph_model, fontsize=14)
```

# ADOM+

```python
nodes = 9 # number of nodes in network
dim = 20 # dim of x
```

## Ring graph

```python
model = methods.ModelTimeVarying(nodes, dim, graph_model='ring')
```

```python
x_f, y_f, z_f, f_err, cons_err, dist = methods.ADOM_PLUS(iters=10000, model=model)
```

```python
plot_results(f_err, cons_err, dist, method='ADOM+', graph_model='Ring graph')
```
